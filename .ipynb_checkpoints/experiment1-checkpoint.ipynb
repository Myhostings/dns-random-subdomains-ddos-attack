estimate the real cardinality of each domain

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('test').getOrCreate()
df = spark.read.csv('data/dns-requests.csv', mode="DROPMALFORMED",inferSchema=True, header = True)
df.createOrReplaceTempView("dns_requests")
distinct_subdomain_dns_requests_df = spark.sql("Select domain, count(distinct subdomain) as distinct_subdomain from dns_requests group by domain order by distinct_subdomain desc limit 100")
distinct_subdomain_dns_requests_pd = distinct_subdomain_dns_requests_df.toPandas()
distinct_subdomain_dns_requests_df.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(range(100), distinct_subdomain_dns_requests_pd['distinct_subdomain'])
plt.title("Real cardinality of top 100 domains")
plt.xlabel("Domain index")
plt.ylabel("Cardinality")
plt.show()
```

Evaluation of the algorithm cardinality estimation for different Ks. 

```python
from algorithm import DistinctHeavyHitters
ks = [200, 500, 1000, 5000, 10000]
dwsHHs = [DistinctHeavyHitters(k=k) for k in ks]
from smart_open import open
first_run = True
for line in open('data/dns-requests.csv.gz'):
    if first_run:
        first_run = False
        continue
    values = line.rstrip().split(',')
    if(len(values)==2):
        subdomain, domain = values
        for dwsHH in dwsHHs:     
            dwsHH.update(subdomain, domain)
```

```python
top_domains = distinct_subdomain_dns_requests_pd["domain"]
card_estimation = {}
for i in range(5):
    for item in dwsHHs[i].count():
        card_estimation[item[0]] = item[1]
    top_domains_card_estimation = []
    for domain in top_domains:
        if domain in card_estimation:    
            top_domains_card_estimation.append(card_estimation[domain])
        else:
            top_domains_card_estimation.append(-1000000)
    plt.plot(range(100), distinct_subdomain_dns_requests_pd['distinct_subdomain'], label='real cardnality')
    plt.plot(range(100), top_domains_card_estimation, label='estimated cardnality')
    plt.title("Real vs Estimated cardinality with k = %d" % ks[i])
    plt.xlabel("Domain index")
    plt.ylabel("Cardinality")
    plt.legend();
    plt.show()
```

```python
error = 0 
for item in dwsHHs[4].count():
    card_estimation[item[0]] = item[1]
top_domains_card_estimation = []
for domain in top_domains:
    top_domains_card_estimation.append(card_estimation[domain])
for i in range(100):
    error += (abs(distinct_subdomain_dns_requests_pd['distinct_subdomain'][i] - top_domains_card_estimation[i]))/distinct_subdomain_dns_requests_pd['distinct_subdomain'][i]
error / 100
```

```python
import os
import psutil
process = psutil.Process(os.getpid())
dwsHH = DistinctHeavyHitters(k=10000)
first_run = True
m1 = process.memory_info().rss
print(m1)
for line in open('data/dns-requests.csv.gz'):
    if first_run:
        first_run = False
        continue
    values = line.rstrip().split(',')
    if(len(values)==2):
        subdomain, domain = values
        dwsHH.update(subdomain, domain)
m2 = process.memory_info().rss
print(m2)
print(m2 - m1)
```
